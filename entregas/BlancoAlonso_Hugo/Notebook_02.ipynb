Ejemplo:

````r
# OJO: R fuerza a que todos los datos sean del mismo tipo (Coerción)
mezcla <- c(1, 2, "tres")
mezcla
class(mezcla)
# Resultado: "character" (convierte todo a texto)
````

## 2) Operaciones vectorizadas

En R, las operaciones matemáticas se aplican **directamente a vectores completos**.

Si sumas, restas, multiplicas o divides un número por un vector,
R aplica la operación a **cada elemento automáticamente**, sin necesidad de bucles.

Esto se llama **operar de forma vectorizada**.

````r
edades <- c(16, 17, 18, 19, 20)

edades + 1        # suma 1 a cada edad
edades * 2        # multiplica por 2 cada edad
edades / 2        # divide cada edad entre 2
````

---

### ¿Por qué esto es importante en IA?

En Inteligencia Artificial trabajamos con:

* listas de datos,
* sensores,
* resultados de modelos,
* predicciones.

La vectorización permite:

* código más corto,
* menos errores,
* mejor rendimiento con muchos datos.

---

### Comparación rápida con Python

En Python, sin librerías como NumPy, necesitarías un bucle:

```python
for i in edades:
    print(i + 1)
```

En R, **no hace falta**.

💡 **Idea clave**

> En R, primero pensamos en vectores y después (si hace falta) en bucles.

## 3) Funciones básicas con vectores

R incluye muchas **funciones integradas** para trabajar con vectores.
Estas funciones se usan constantemente en **análisis de datos y Machine Learning** para describir y resumir información.

Las más habituales son:

* `length(x)` → número de elementos del vector
* `sum(x)` → suma de los valores
* `mean(x)` → media (promedio)
* `min(x)` → valor mínimo
* `max(x)` → valor máximo

---

### Ejemplo




````r
length(edades)
sum(edades)
mean(edades)
min(edades)
max(edades)
````

---

### ¿Por qué esto es importante en IA?

Estas funciones permiten:

* entender cómo se comportan los datos,
* detectar valores extremos,
* resumir resultados de modelos,
* comparar distintas ejecuciones de un algoritmo.

En Machine Learning, **antes de entrenar un modelo**, siempre se analizan los datos con este tipo de funciones.



### Datos incompletos (muy común)

En Data Science real, los datos vienen sucios. Python usa **NaN** o **None**. R usa **NA**. 



````r
# Datos sucios (muy común en ML)
datos_sucios <- c(10, 20, NA, 40)

mean(datos_sucios) 
# Resultado: NA (R es conservador, si hay un desconocido, el resultado es desconocido)

# La solución:
mean(datos_sucios, na.rm = TRUE)
````


## 4) Acceder a elementos (indexado)

En R, los **índices empiezan en 1**, no en 0 como en Python.
Este es uno de los errores más comunes al empezar.

⚠️ **Recuerda**:

* primer elemento → posición **1**
* último elemento → posición `length(x)`

---

### Ejemplos básicos


````r
edades <- c(16, 17, 18, 19, 20)

edades[1]
edades[3]
edades[2:4]
edades[c(1, 5)] 
````

### Comparación rápida con Python

* En Python: `x[0]` → primer elemento
* En R: `x[1]` → primer elemento

💡 **Regla mental útil**

> En R, contar empieza en 1.

---

### ¿Por qué esto importa en IA?

Al trabajar con datos y modelos:

* seleccionamos subconjuntos de datos,
* inspeccionamos valores concretos,
* analizamos resultados extremos (mínimos, máximos).

Un error en el indexado puede hacer que analices **el dato equivocado**.

---


### Un detalle importante en R

En R, usar un índice negativo **no significa lo mismo que en Python**.

```r
x[-1]
```

Esto significa:

> “devuélveme todos los elementos **excepto** el primero”.

En Python, `x[-1]` devuelve el último elemento.
En R, **no**.

````r
edades[-1] # todas excepto la primera
````

##  Ejemplo: función para evaluar un modelo

Supongamos que tenemos la **precisión** de un modelo y queremos devolver un mensaje automático.

###  En R, una función se crea con `function()`





````r
evaluar_modelo <- function(accuracy) {
  if (accuracy >= 0.80) {
    return("Modelo aceptable")
  } else {
    return("Modelo no aceptable")
  }
}
````

Usamos la función así:

````r
resultado <- evaluar_modelo(0.83)
print(resultado)
````

---

##  Qué está pasando aquí

* `evaluar_modelo` → nombre de la función
* `accuracy` → parámetro de entrada
* `{}` → cuerpo de la función
* `return()` → valor que devuelve la función

> 💡 Una función es **una máquina**: entra un valor, sale un resultado.

---

##  Comparación directa con Python 

### En Python haríamos algo así:

```python
def evaluar_modelo(accuracy):
    if accuracy >= 0.80:
        return "Modelo aceptable"
    else:
        return "Modelo no aceptable"
```

### Diferencias clave R vs Python

| Aspecto       | R                     | Python      |
| ------------- | --------------------- | ----------- |
| Palabra clave | `function()`          | `def`       |
| Asignación    | `<-`                  | `=`         |
| Bloques       | `{}`                  | indentación |
| Indentación   | recomendada           | obligatoria |
| `return`      | opcional (pero clara) | habitual    |


En R, **las llaves mandan**, no los espacios.

En análisis de datos y Machine Learning es importante detectar **valores anómalos** (outliers), ya que pueden afectar negativamente al rendimiento de un modelo.

En este ejercicio analizaremos los **errores de un modelo**.

---

### Datos iniciales

0.12, 0.15, 0.11, 0.14, 0.13, 0.16, 0.12, 0.14, 0.15, 0.13, 0.11, 0.12, 0.14, 0.13, 0.15, 0.16, 0.12, 0.14, 0.13, 0.11, 0.18, 0.17, 0.16, 0.15, 0.14, 0.13, 0.45, 0.50, 0.60


Cada valor representa el error cometido por un modelo en distintas observaciones.

---

### 🔧 Enunciado

1. Calcula la **media** del vector `errores`.
2. Filtra los errores que sean **mayores que la media**.
3. Crea una función que muestre:

   * la media de los errores,
   * los errores considerados altos,
   * cuántos errores son mayores que la media.

---

### Pistas

* Usa `mean()` para calcular la media.
* Usa comparaciones vectorizadas (`>`) para detectar errores altos.
* Usa `length()` para contar cuántos valores cumplen una condición.
* Usa indexado lógico con `[]` para filtrar.


````r
# TODO: completa
errores <- c(0.12, 0.15, 0.11, 0.14, 0.13, 0.16, 0.12, 0.14, 0.15, 0.13, 0.11, 0.12, 0.14, 0.13, 0.15, 0.16, 0.12, 0.14, 0.13, 0.11, 0.18, 0.17, 0.16, 0.15, 0.14, 0.13, 0.45, 0.50, 0.60)

media <- mean(errores)
mayor_media <- errores[errores > media]

fun <-function(x) {
    media <- mean(x)
    print(media)
    mayores <- x[x > media]
    print(mayores)
    print(length(mayores))
}

media
mayor_media
fun(errores)
````
